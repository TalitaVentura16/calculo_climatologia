```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'serif'

# Carregar dados do Excel
amazonia = '/home/talita/Documentos/calculo_climatologia/estudoEstacaoSeca/Dados/Amazonia_pentada_areaDesmatada2.xlsx'
df = pd.read_excel(amazonia)

# Transformar o DataFrame para formato longo (long format)
df_long = pd.melt(df, id_vars=["Pentada"], var_name="Ano", value_name="Precipitao")

# Remover linhas onde "Ano" no seja um nmero (exemplo: "Mdia")
df_long = df_long[pd.to_numeric(df_long["Ano"], errors="coerce").notna()]

# Converter a coluna "Ano" para numrico
df_long["Ano"] = pd.to_numeric(df_long["Ano"])

# Converter a coluna "Pentada" para numrico, removendo a palavra ""
df_long["Pentada"] = df_long["Pentada"].str.replace("", "").astype(int)

# Filtrar os dados de 1981 a 2023
df_filtered = df_long[(df_long["Ano"] >= 1981) & (df_long["Ano"] <= 2023)]

# Lista para armazenar informaes da estao seca e chuvosa
season_info = []

# Loop pelos anos para calcular A(day) e identificar incio e fim das estaes
for year in range(1981, 2024):  # Ajustado para incluir 2023
    year_data = df_filtered[df_filtered["Ano"] == year].copy()

    if year_data.empty:
        print(f"Ano {year} no possui dados suficientes para clculo.")
        continue

    # Calcular a mdia anual diria
    R_mean = year_data["Precipitao"].mean()

    # Calcular A(day) acumulando precipitao e subtraindo a mdia acumulada
    year_data["A_day"] = year_data["Precipitao"].cumsum() - R_mean * year_data["Pentada"]

    # Identificar os pontos de mudana em A_day (incio e fim da estao chuvosa)
    try:
        min_index = year_data["A_day"].idxmin()  # ndice do mnimo de A(day)
        max_index = year_data["A_day"].idxmax()  # ndice do mximo de A(day)

        # Garantir que o incio (mnimo) vem antes do fim (mximo)
        if min_index < max_index:
            inicio_chuva = year_data.loc[min_index, "Pentada"]
            fim_chuva = year_data.loc[max_index, "Pentada"]
        else:
            inicio_chuva = year_data.loc[max_index, "Pentada"]
            fim_chuva = year_data.loc[min_index, "Pentada"]
    except IndexError:
        print(f"Ano {year} no possui dados suficientes para identificar estao chuvosa.")
        continue

    # Calcular durao da estao seca como complementares  estao chuvosa
    pentadas_total = 73  # Total de pentadas no ano
    duracao_chuva = fim_chuva - inicio_chuva
    duracao_seca = pentadas_total - duracao_chuva

    # Adicionar informaes ao dataframe
    season_info.append({
        "Ano": year,
        "Pentada_Incio_Chuva": inicio_chuva,
        "Pentada_Fim_Chuva": fim_chuva,
        "Durao_Chuva": duracao_chuva,
        "Durao_Seca": duracao_seca
    })

# Transformar as informaes da estao em DataFrame
season_df = pd.DataFrame(season_info)

# Salvar as informaes em um arquivo Excel
#output_excel_path = "/home/talita/Documentos/calculo_climatologia/estudoEstacaoSeca/Resultados/Estacoes_Resumo.xlsx"
#season_df.to_excel(output_excel_path, index=False)

# Plotar grficos (mesmo mtodo que antes, mas adaptado para o novo conceito)
output_path = "/home/talita/Documentos/calculo_climatologia/estudoEstacaoSeca/Resultados/AreaDesmatada3_analise.pdf"
fig, axes = plt.subplots(nrows=22, ncols=2, figsize=(20, 80))
axes = axes.flatten()

for i, year in enumerate(range(1981, 2024)):
    ax = axes[i]
    year_data = df_filtered[df_filtered["Ano"] == year]

    if year_data.empty:
        ax.set_title(f"Dados faltantes - {year}")
        continue

    ax.plot(year_data["Pentada"], year_data["Precipitao"], label="Precipitao", marker='o', color='blue')

    if "A_day" in year_data.columns:
        ax.plot(year_data["Pentada"], year_data["A_day"], label="A(day)", color="red", linestyle="--")

    dry_year = season_df[season_df["Ano"] == year]
    if not dry_year.empty:
        ax.axvline(x=dry_year["Pentada_Incio_Chuva"].values[0], color="green", linestyle="--", label="Incio da Estao Chuvosa")
        ax.axvline(x=dry_year["Pentada_Fim_Chuva"].values[0], color="purple", linestyle="--", label="Fim da Estao Chuvosa")

    ax.set_title(f"Precipitao para o Ano de {year}", fontweight='bold', fontsize=14)
    ax.set_xlabel("Pentada")
    ax.set_ylabel("Precipitao (mm)", fontweight='bold')
    ax.legend()
    ax.grid(alpha=0.5)

plt.tight_layout()
plt.savefig(output_path, format='pdf')
plt.show()

# Exibir as informaes da estao
print(season_df)
```

```python
# Clculos estatsticos sobre a durao da estao seca considerando todo o perodo fora da estao chuvosa
def analyze_season_data(season_df):
    # Durao mdia da estao seca
    mean_duration = season_df["Durao_Seca"].mean()

    # Durao mnima e mxima
    min_duration = season_df["Durao_Seca"].min()
    max_duration = season_df["Durao_Seca"].max()

    # Ano com menor e maior durao
    min_year = season_df.loc[season_df["Durao_Seca"].idxmin(), "Ano"]
    max_year = season_df.loc[season_df["Durao_Seca"].idxmax(), "Ano"]

    # Aumento percentual da durao ao longo dos anos
    initial_duration = season_df.iloc[0]["Durao_Seca"]
    final_duration = season_df.iloc[-1]["Durao_Seca"]
    percentage_change = ((final_duration - initial_duration) / initial_duration) * 100

    # Exibir resultados no terminal
    print("\n--- Anlise da Durao da Estao Seca ---")
    print(f"Durao Mdia (pentadas): {mean_duration:.2f}")
    print(f"Durao Mnima (pentadas): {min_duration} (Ano: {min_year})")
    print(f"Durao Mxima (pentadas): {max_duration} (Ano: {max_year})")
    print(f"Aumento Percentual ao longo dos anos: {percentage_change:.2f}%\n")

    # Retornar os resultados (caso precise us-los depois)
    return {
        "Durao Mdia (pentadas)": mean_duration,
        "Durao Mnima (pentadas)": min_duration,
        "Ano Mnimo": min_year,
        "Durao Mxima (pentadas)": max_duration,
        "Ano Mximo": max_year,
        "Aumento Percentual (%)": percentage_change
    }

# Atualizar clculo da estao seca para incluir todos os dias fora da estao chuvosa
for year in range(1981, 2024):  # Ajustado para incluir 2023
    year_data = df_filtered[df_filtered["Ano"] == year].copy()

    if year_data.empty:
        print(f"Ano {year} no possui dados suficientes para clculo.")
        continue

    # Calcular a mdia anual diria
    R_mean = year_data["Precipitao"].mean()

    # Calcular A(day) acumulando precipitao e subtraindo a mdia acumulada
    year_data["A_day"] = year_data["Precipitao"].cumsum() - R_mean * year_data["Pentada"]

    # Identificar os pontos de mudana em A_day (incio e fim da estao seca)
    try:
        min_index = year_data["A_day"].idxmin()  # ndice do mnimo de A(day)
        max_index = year_data["A_day"].idxmax()  # ndice do mximo de A(day)

        # Garantir que o incio (mnimo) vem antes do fim (mximo)
        if min_index < max_index:
            inicio_chuvosa = year_data.loc[min_index, "Pentada"]
            fim_chuvosa = year_data.loc[max_index, "Pentada"]
        else:
            inicio_chuvosa = year_data.loc[max_index, "Pentada"]
            fim_chuvosa = year_data.loc[min_index, "Pentada"]

        # Determinar os dias fora da estao chuvosa como pertencentes  estao seca
        dias_seca = 73 - (fim_chuvosa - inicio_chuvosa)  # Total de pentadas no ano  73
    except IndexError:
        print(f"Ano {year} no possui dados suficientes para identificar estao chuvosa.")
        continue

    # Adicionar informaes ao dataframe
    season_info.append({
        "Ano": year,
        "Pentada_Incio_Chuvosa": inicio_chuvosa,
        "Pentada_Fim_Chuvosa": fim_chuvosa,
        "Durao_Seca": dias_seca
    })

# Transformar as informaes da estao em DataFrame
season_df = pd.DataFrame(season_info)

# Realizar anlises e exibir no terminal
analysis_results = analyze_season_data(season_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
import statsmodels.api as sm

rcParams['font.family'] = 'serif'

# Carregar dados do Excel
amazonia = '/home/talita/Documentos/calculo_climatologia/estudoEstacaoSeca/Dados/Amazonia_pentada_areaDesmatada2.xlsx'
df = pd.read_excel(amazonia)

# Transformar o DataFrame para formato longo (long format)
df_long = pd.melt(df, id_vars=["Pentada"], var_name="Ano", value_name="Precipitao")

# Remover linhas onde "Ano" no seja um nmero (exemplo: "Mdia")
df_long = df_long[pd.to_numeric(df_long["Ano"], errors="coerce").notna()]

# Converter a coluna "Ano" para numrico
df_long["Ano"] = pd.to_numeric(df_long["Ano"])

# Converter a coluna "Pentada" para numrico, removendo a palavra ""
df_long["Pentada"] = df_long["Pentada"].str.replace("", "").astype(int)

# Filtrar os dados de 1981 a 2023
df_filtered = df_long[(df_long["Ano"] >= 1981) & (df_long["Ano"] <= 2023)]

# Criar precipitation_df baseado nos dados filtrados
precipitation_df = df_filtered.groupby("Ano")["Precipitao"].sum().reset_index()
precipitation_df.rename(columns={"Precipitao": "Total_Precipitao"}, inplace=True)

# Funo para anlise de precipitao
def analyze_precipitation_data(precipitation_df):
    # Precipitao mdia anual
    mean_precipitation = precipitation_df["Total_Precipitao"].mean()

    # Precipitao mnima e mxima
    min_precipitation = precipitation_df["Total_Precipitao"].min()
    max_precipitation = precipitation_df["Total_Precipitao"].max()

    # Ano com menor e maior precipitao
    min_year = precipitation_df.loc[precipitation_df["Total_Precipitao"].idxmin(), "Ano"]
    max_year = precipitation_df.loc[precipitation_df["Total_Precipitao"].idxmax(), "Ano"]

    # Aumento percentual da precipitao ao longo dos anos
    initial_precipitation = precipitation_df.iloc[0]["Total_Precipitao"]
    final_precipitation = precipitation_df.iloc[-1]["Total_Precipitao"]
    percentage_change = ((final_precipitation - initial_precipitation) / initial_precipitation) * 100

    # Exibir resultados no terminal
    print("\n--- Anlise da Precipitao Anual rea Desmatada 2---")
    print(f"Precipitao Mdia Anual (mm): {mean_precipitation:.2f}")
    print(f"Precipitao Mnima Anual (mm): {min_precipitation} (Ano: {min_year})")
    print(f"Precipitao Mxima Anual (mm): {max_precipitation} (Ano: {max_year})")
    print(f"Aumento Percentual ao longo dos anos: {percentage_change:.2f}%\n")

    # Retornar os resultados
    return {
        "Precipitao Mdia Anual (mm)": mean_precipitation,
        "Precipitao Mnima Anual (mm)": min_precipitation,
        "Ano Mnimo": min_year,
        "Precipitao Mxima Anual (mm)": max_precipitation,
        "Ano Mximo": max_year,
        "Aumento Percentual (%)": percentage_change
    }

# Realizar a anlise de precipitao
precipitation_analysis = analyze_precipitation_data(precipitation_df)

# Exibir o DataFrame de resultados
print(precipitation_analysis)
```

```python
import matplotlib.pyplot as plt

# Funo para combinar os grficos de precipitao e durao da estao seca
def plot_combined_trend(precipitation_df, season_df):
    # Criar a figura e os eixos
    fig, ax1 = plt.subplots(figsize=(12, 7))

    # Grfico da precipitao (eixo esquerdo)
    ax1.plot(
        precipitation_df["Ano"], precipitation_df["Total_Precipitao"], 
        label="Precipitao Anual", color='purple', marker='o'
    )
    mean_precipitation = precipitation_df["Total_Precipitao"].mean()
    ax1.axhline(mean_precipitation, color='purple', linestyle='--', label=f"Precipitao Mdia ({mean_precipitation:.2f} mm)")
    ax1.set_xlabel("Ano", fontsize=12, fontweight='bold')
    ax1.set_ylabel("Precipitao Anual (mm)", fontsize=12, fontweight='bold', color='purple')
    ax1.tick_params(axis='y', labelcolor='purple')
    ax1.grid(alpha=0.5)

    # Grfico da durao da estao seca (eixo direito)
    ax2 = ax1.twinx()
    ax2.plot(
        season_df["Ano"], season_df["Durao_Seca"], 
        label="Durao da Estao Seca", color='blue', marker='o'
    )
    mean_duration = season_df["Durao_Seca"].mean()
    ax2.axhline(mean_duration, color='blue', linestyle='--', label=f"Durao Mdia ({mean_duration:.2f} pentadas)")
    ax2.set_ylabel("Durao da Estao Seca (pentadas)", fontsize=12, fontweight='bold', color='blue')
    ax2.tick_params(axis='y', labelcolor='blue')

    # Configurar ticks do eixo x para mostrar todos os anos
    all_years = sorted(precipitation_df["Ano"].unique())  # Listar todos os anos nicos
    ax1.set_xticks(all_years)  # Definir os anos como ticks no eixo x
    ax1.set_xticklabels(all_years, rotation=45, fontsize=8)  # Rotacionar os anos para melhor legibilidade

    # Combinar legendas dos dois eixos
    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax2.get_legend_handles_labels()
    ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc="upper left", fontsize=10)

    # Ttulo do grfico
    plt.title("Precipitao Anual e Durao da Estao Seca - rea Desmatada 2", fontsize=16, fontweight='bold')

    # Ajustar layout
    plt.tight_layout()

    # Mostrar o grfico
    plt.show()

# Chamar a funo para plotar os grficos combinados
plot_combined_trend(precipitation_df, season_df)
```
